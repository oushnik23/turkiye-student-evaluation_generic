```python
data=dataset.iloc[:,5:]
```

```python
data
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=None)
data=pca.fit_transform(data)
explained_variance=pca.explained_variance_ratio_
```

```python
explained_variance
```

```python
from sklearn.cluster import KMeans
```

```python
WCSS=[]
```

```python
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(data)
  WCSS.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),WCSS)
plt.title("Elbow method")
plt.xlabel("Number of Cluster")
plt.ylabel('WCSS')
```

```python
kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(data)
```

```python
y_kmeans
```

```python
plt.scatter(data[y_kmeans==0,0],data[y_kmeans==0,1],s=100,c='Red',label='cluster1')
plt.scatter(data[y_kmeans==1,0],data[y_kmeans==1,1],s=100,c='cyan',label='cluster2')
plt.scatter(data[y_kmeans==2,0],data[y_kmeans==2,1],s=100,c='Green',label='Cluster3')
plt.scatter(data[y_kmeans==3,0],data[y_kmeans==3,1],s=100,c='Blue',label='Cluster4')
plt.title('Cluster of Students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
```

```python
df_cluster=pd.DataFrame(y_kmeans,columns=['Cluster'])
```

```python
df_cluster
```

```python
dataset=dataset.join(df_cluster)
```
